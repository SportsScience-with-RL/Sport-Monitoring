```python
marqueurs = ['marqueur1', 'marqueur2', 'marqueur3', 'marqueur4', 'marqueur5', 'marqueur6', 'marqueur7', 'marqueur8', 
             'marqueur9', 'marqueur10', 'marqueur11', 'marqueur12', 'marqueur13', 'marqueur14', 'marqueur15',
             'marqueur16', 'marqueur17', 'marqueur18']
```

### Dendrogramme données standardisées

```python
X = df.copy()
#Séléctionner une période de plusieurs semaines pour l'analyse (ici 5 semaines)
X = X[(X['numsemaine'] >= 2042) & (X['numsemaine'] <= 2046)]
X_features = X.iloc[:,5:]
```

```python
z_scaler = StandardScaler()
X_z = z_scaler.fit_transform(X_features)

hac_dend = linkage(X_z, method='ward', metric='euclidean')
plt.figure(figsize=(17,17))
dendrogram(hac_dend)
plt.show()
```

##### Obtention donc de 3 classes (visualisation 4, mais la quatrième classe est minime)

### Classification hiérarchique ascendante à 4 clusters (la CHA doit tout de même être exécutée avec un paramètre de 4 clusters)

```python
model = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
model.fit(X_z)
labels = model.labels_

X['Cluster_Label'] = labels
X['Cluster_Label'].value_counts()
```

### Composition des clusters

```python
X['Total_sportif'] = X.groupby('sportif').numsemaine.transform('count')

X['Cluster_1'] = X.groupby('sportif')['Cluster_Label'].transform(lambda x: (x==0).sum())
X['Cluster_2'] = X.groupby('sportif')['Cluster_Label'].transform(lambda x: (x==1).sum())
X['Cluster_3'] = X.groupby('sportif')['Cluster_Label'].transform(lambda x: (x==2).sum())
X['Cluster_4'] = X.groupby('sportif')['Cluster_Label'].transform(lambda x: (x==3).sum())

X['% Cluster_1'] = ((X['Cluster_1'] / X['Total_sportif']) * 100).round()
X['% Cluster_2'] = ((X['Cluster_2'] / X['Total_sportif']) * 100).round()
X['% Cluster_3'] = ((X['Cluster_3'] / X['Total_sportif']) * 100).round()
X['% Cluster_4'] = ((X['Cluster_4'] / X['Total_sportif']) * 100).round()

X_ct = X[['sportif', '% Cluster_1', '% Cluster_2', '% Cluster_3','% Cluster_4', 'Cluster_Label']]

agg_func = {}

for c in X_ct.columns[1:]:
    if c != 'Cluster_Label':
        agg_func[c] = 'median'
    else:
        agg_func[c] = pd.Series.mode

#Etablir un tableau récapitulatif avec la sensibilité (%) de chaque joueur pour chaque cluster et 
#l'affiliation au cluster correspondant le plus
X_recap = X_ct.groupby('sportif').agg(agg_func)

X_recap['Cluster_Label'].replace({0:'Cluster 1', 1:'Cluster 2', 2:'Cluster 3', 3:'Cluster 4'}, inplace=True)
X_recap.sort_values(by='Cluster_Label')
```

##### Comme vu sur le dendrogramme, le quatrième cluster est minime, on ne retrouve aucune affiliation à celui-ci

### Tableau descriptif des clusters (médianes et différences par cluster)

```python
X['Cluster_Label'].replace({0:'Cluster 1', 1:'Cluster 2', 2:'Cluster 3', 3:'Cluster 4'}, inplace=True)
df_recap_cluster = pd.DataFrame(columns=['Cluster 1', 'Cluster 2', 'Cluster 3', 'Global'], index=marqueurs)

for col in df_recap_cluster.columns:
    for m in marqueurs:
        diff = X.loc[X['Cluster_Label']==col, m].median() - X[m].median()
        df_recap_cluster.loc[m,col] = [np.round(diff,2), X.loc[X['Cluster_Label']==col, m].median()]
        df_recap_cluster.loc[m, 'Global'] = X[m].median()
```

```python
df_recap_cluster.rename({'Cluster 1': 'Cluster 1 [diff, médiane]',
                         'Cluster 2': 'Cluster 2 [diff, médiane]',
                         'Cluster 3': 'Cluster 3 [diff, médiane]', 
                         'Global': 'Global (médiane)'}, axis='columns', inplace=True)

df_recap_cluster.style.set_properties(**{'text-align': 'center'})
```

##### On voit que le cluster 2 a tendance à saisir en dessous de la médiane du groupe entier, le cluster 1 au dessus

### PCA pour visualiser les clusters
#### On vérifie la pertinence des variables avec le test de Kaiser-Meyer-Olkin (KMO)

```python
kmo_all, kmo_model = calculate_kmo(X_features)
print('Overall KMO : {}'.format(kmo_model))
pd.DataFrame(index=marqueurs, data=kmo_all, columns=['KMO'])
```

##### Le KMO est faible pour 'marqueur16' et 'marqueur18'. Le KMO est à la limite de l'acceptable pour 'marqueur3' et 'marqueur4', on va garder tout de même conserver 'marqueur4. Le KMO général est excellent

```python
X_pca = X_features.drop(columns=['marqueur3', 'marqueur16', 'marqueur18'])
```

```python
pca_scaler = StandardScaler()
X_pca_scaled = pca_scaler.fit_transform(X_pca)

pca = PCA(n_components=5)
X_new = pca.fit_transform(X_pca_scaled)
```

```python
X_new_final = pd.DataFrame(X_new, columns=['PC1','PC2','PC3','PC4','PC5'])
X_new_final['Cluster'] = labels
X_new_final['Cluster'].replace({0:'Cluster 1', 1:'Cluster 2', 2:'Cluster 3', 3:'Cluster 4'}, inplace=True)
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=X_new_final, x='PC2', y='PC1', hue='Cluster', palette='viridis_r')
```

##### On distingue bien nos 3 clusters et on peut constater que le cluster 1 saisie au-dessus des autres

### Distributions des différents marqueurs selon les clusters 

```python
X_plot = X[X['Cluster_Label'] != 'Cluster 4']

def plot(Marqueur):
    fig = ff.create_distplot([X_plot.loc[X_plot['Cluster_Label']==c, Marqueur] for c in np.sort(X_plot['Cluster_Label'].unique())],
                             np.sort(X_plot['Cluster_Label'].unique()),
                             show_rug=False, bin_size=1, show_hist=False, colors=['#191970', '#2E8B57', '#6BF3A8'])
    fig.update_layout(template='plotly_white')
    return fig

widgets.interact(plot, Marqueur=marqueurs)
```
